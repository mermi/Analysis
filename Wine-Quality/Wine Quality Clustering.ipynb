```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
```

```python
X = wines[['residual sugar', 'chlorides']].as_matrix()
labels = kmeans.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels);
```

```python
def make_shit(feature_a, feature_b, df):
    X = df[[feature_a, feature_b]].as_matrix()
    kmeans = KMeans(n_clusters=2, random_state=42)
    labels = kmeans.fit_predict(X)
    plt.scatter(X[:, 0], X[:, 1], c=labels);
    plt.show()
```

```python
import itertools

combs = itertools.combinations(wines.columns, 2)

for e in combs:
    if not any([x == 'color' for x in e]):
        print(e)
        make_shit(e[0], e[1], wines)
```
